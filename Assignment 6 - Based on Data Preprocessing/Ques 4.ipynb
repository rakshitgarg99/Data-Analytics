### 4. There are 10 columns in the dataset with non-numeric values. Convert these values to numeric values using following scheme:

### (i) For “num_doors” and “num_cylinders”: convert words (number names) to figures for e.g., two to 2

```python
val_doors = df['num_doors'].unique()
val_cylndr = df['num_cylinders'].unique()
```

```python
help_dict = {
    "two": 2,
    "three": 3,
    "four": 4,
    "five": 5,
    "six": 6,
    "eight": 8,
    "twelve": 12,
    "Nan": None
}
```

```python
for i in range(len(df)):
    df['num_doors'][i] = help_dict[df['num_doors'][i]]
    df['num_cylinders'][i] = help_dict[df['num_cylinders'][i]]

df['num_doors'] = pd.to_numeric(df['num_doors'], errors='coerce')
df['num_cylinders'] = pd.to_numeric(df['num_cylinders'], errors='coerce')
```

### (ii) For "body_style", "drive_wheels": use dummy encoding scheme

```python
df2 = pd.get_dummies(df['body_style'])
df = df.drop(['body_style'],axis=1)
df = pd.concat([df, df2], axis=1)


df2 = pd.get_dummies(df['drive_wheels'])
df = df.drop(['drive_wheels'],axis=1)
df = pd.concat([df, df2], axis=1)
df.head()
```

### (iii) For “make”, “aspiration”, “engine_location”,fuel_type: use label encoding scheme

```python
label_encoder = preprocessing.LabelEncoder()
df['make']= label_encoder.fit_transform(df['make'])
df['aspiration']= label_encoder.fit_transform(df['aspiration'])
df['engine_location']= label_encoder.fit_transform(df['engine_location'])
df['fuel_type']= label_encoder.fit_transform(df['fuel_type'])
```

### (iv) For fuel_system: replace values containing string pfi to 1 else all values to 0.
### (v) For engine_type: replace values containing string ohc to 1 else all values to 0.

```python
df['fuel_system'] = [1 if i.find('pfi')==1 else 0 for i in list(df['fuel_system'])]
df['engine_type'] = [1 if i.find('ohc')==1 else 0 for i in list(df['engine_type'])]
```

## 2. Replace all NaN values with central tendency imputation. Drop the rows with NaN values in price column

```python
idx = df[df["price"]=="Nan"].index
print(idx) # price index with nan values
df = df.drop(idx, axis=0)
df.reset_index(inplace=True)
df = df.drop(['index'],axis=1)
df
```

```python
nan_cols = [i for i in df.columns if len(df[df[i]=="Nan"])!=0]
print(nan_cols)
print(df[nan_cols].dtypes)

for i in nan_cols:
    df[i] = pd.to_numeric(df[i], errors='coerce') # change nan_cols with object type to numeric

# print(df.mean()) # mean of all columns
# df = df.fillna(df.mean())

df['normalized_losses'] = df['normalized_losses'].astype('float64')
df['normalized_losses'].fillna(df['normalized_losses'].median(), inplace = True)
df['normalized_losses'] = df['normalized_losses'].astype('int64')

df['num_doors'] = df['num_doors'].astype('category')
df['num_doors'].fillna(df['num_doors'].mode().iloc[0], inplace = True)

df['bore'] = df['bore'].astype('float64')
df['bore'].fillna(df['bore'].mean(), inplace = True)

df['stroke'] = df['stroke'].astype('float64')
df['stroke'].fillna(df['stroke'].mean(), inplace = True)

df['horsepower'] = df['horsepower'].astype('float64')
df['horsepower'].fillna(df['horsepower'].mean(), inplace = True)
df['horsepower'] = df['horsepower'].astype('int64')

df['peak_rpm'] = df['peak_rpm'].astype('float64')
df['peak_rpm'].fillna(df['peak_rpm'].median(), inplace = True)
df['peak_rpm'] = df['peak_rpm'].astype('int64')


df.head()
```
