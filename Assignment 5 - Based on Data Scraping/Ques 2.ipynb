Ques2. - The page contains information about movies released in the year 2020 sorted according to number
of votes. Scrape the following information for the first 1000 movies.

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
table = []
```

```python
i=1
while(i<=1000):
    url = 'https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page='+str(i)
    i=i+50
    # print(url)
    r = requests.get(url)
    soup = bs(r.content, 'html.parser')
    # print(soup)
    for movie in soup.findAll('div', {'class': 'lister-item mode-advanced'}):
        row = []
        movie_name = movie.find('h3', {'class':'lister-item-header'}).a.text
        # print(movie_name)
        row.append(movie_name)
        imdb_rating = movie.find('div', {'class':'inline-block ratings-imdb-rating'}).strong.text
        # print(imdb_rating)
        row.append(imdb_rating)
        description = movie.findAll('p', {'class':'text-muted'})[1].text.replace('\n',"")
        if description != 'Add a Plot':
            row.append(description)
        else:
            row.append("None")
        # print(description)
        votes = movie.find('p', {'class':'sort-num_votes-visible'}).findAll('span')[1].text
        # print(votes)
        row.append(votes)
        director = movie.find('p', {'class':''}).text
        if director.find('Director:') != -1:
            director_name = director[director.find('Director:')+9: director.find('|')]
        else:
            director_name = 'None'
        # print(director_name)
        row.append(director_name.replace('\n',""))
        stars = movie.find('p', {'class':''}).text
        if stars.find('Stars:') != -1:
            stars_name = stars[stars.find('Stars:')+6:]
        elif stars.find('Star:') != -1:
            stars_name = stars[stars.find('Star:')+5:]
        else:
            stars_name = 'None'
        # print(stars_name)
        row.append(stars_name.replace('\n',""))

        metascore = movie.find('span', {'class':'metascore'})
        if metascore:
            row.append(metascore.text.replace('\n',""))
        else:
            row.append('None')
        # print(metascore)

        movie_genre = movie.find('span', {'class':'genre'})
        # print(movie_genre)
        if movie_genre:
            row.append(movie_genre.text.replace('\n',""))
        else:
            row.append('None')
        total_duration = movie.find('span', {'class':'runtime'})
        # print(total_duration)
        if total_duration:
            row.append(total_duration.text)
        else:
            row.append('None')
        table.append(row)
```

```python
table
```

```python
df = pd.DataFrame(table,columns=['Movie Name','IMDB Rating','Description','Votes', 'Director(s)','Stars','Metascore','Movie Genre','Total Duration'])
```
