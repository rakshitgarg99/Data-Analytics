Ques1. - For each of the best seller book mentioned on the first two pages of the following link
https://www.amazon.in/gp/bestsellers/books/

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
table = []
```

```python
def get_data(pageno):
    url=('https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageno)+'?ie=UTF8&pg='+str(pageno))
    print(url)
    r=requests.get(url)
    html_content=r.content
    soup=BeautifulSoup(html_content, 'html.parser')
    for books_list in soup.findAll('li', {'class' : 'zg-item-immersion'}):
        book_link = books_list.find('a', {'class':'a-link-normal'}).get('href')
        row = []
        link = 'https://www.amazon.in'+book_link
        print(link)

        browser = webdriver.Chrome("C:/Users/raksh/Downloads/chromedriver_win32/chromedriver.exe")
        browser.get(link)

        book_html = browser.page_source
        # print(book_html)

        book_soup=BeautifulSoup(book_html, 'html.parser')
        # print(book_soup)

        product_detail = book_soup.find('ul', {'class':'a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list'}).text
        # print(product_detail)

        publisher = product_detail[product_detail.find('Publisher')+13:product_detail.find('Language')]
        # print(publisher.replace('\n',""))
        row.append(publisher.replace('\n',""))
        
        if product_detail.find('Board book') != -1:
            language = product_detail[product_detail.find('Language')+12:product_detail.find('Board book')]
        elif product_detail.find('Paperback') != -1:
            language = product_detail[product_detail.find('Language')+12:product_detail.find('Paperback')]
        # print(language.replace('\n',""))
        row.append(language.replace('\n',""))
        
        if product_detail.find('Board book') != -1:
            paperback = product_detail[product_detail.find('Board book')+14:product_detail.find('ISBN-10')]
        elif product_detail.find('Paperback') != -1:
            paperback = product_detail[product_detail.find('Paperback')+14:product_detail.find('ISBN-10')]
        # print(paperback.replace('\n',""))
        row.append(paperback.replace('\n',""))
        
        isbn10 = product_detail[product_detail.find('ISBN-10')+11:product_detail.find('ISBN-13')]
        # print(isbn10.replace('\n',""))
        row.append(isbn10.replace('\n',""))
        
        if product_detail.find('Reading age') != -1:
            isbn13 = product_detail[product_detail.find('ISBN-13')+11:product_detail.find('Reading age')]
        elif product_detail.find('Item Weight') != -1:
            isbn13 = product_detail[product_detail.find('ISBN-13')+11:product_detail.find('Item Weight')]
        # print(isbn13.replace('\n',""))
        row.append(isbn13.replace('\n',""))
        
        weight = product_detail[product_detail.find('Item Weight')+15:product_detail.find('Dimensions')]
        # print(weight.replace('\n',""))
        row.append(weight.replace('\n',""))
        
        dimensions = product_detail[product_detail.find('Dimensions')+15:product_detail.find('Country of Origin')]
        # print(dimensions.replace('\n',""))
        row.append(dimensions.replace('\n',""))


        reviews_list = book_soup.findAll('div', {'class':'a-row a-spacing-small review-data'})
        
        try:
            first_review = reviews_list[0].text.replace('\n',"")[:-9]
        except:
            first_review = 'None'
        
        try:
            second_review = reviews_list[1].text.replace('\n',"")[:-9]
        except:
            second_review = 'None'
        
        try:
            third_review = reviews_list[2].text.replace('\n',"")[:-9]
        except:
            third_review = 'None'


        # print(first_review)
        # print(second_review)
        # print(third_review)
        row.append(first_review)
        row.append(second_review)
        row.append(third_review)

        table.append(row)
```

```python
get_data(1)
```

```python
get_data(2)
```

```python
table
```

```python
df = pd.DataFrame(table,columns=['Publisher','Language','Paperback','ISBN-10', 'ISBN-13','Weight','Dimensions','First Customer Review', 'Second Customer Review', 'Third Customer Review'])
```
